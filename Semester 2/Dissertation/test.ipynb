```python
import threading
import pandas as pd
import time 
from instagrapi import Client

# Load the dataset
proxies = pd.read_csv("http_socks5_socks4_proxies.csv")
print(proxies.info())
# Select the required columns
extracted_proxies = proxies[["ip", "port", "protocol"]]

# Format the data as "protocol://ip:port"
formatted_proxies = extracted_proxies.apply(lambda row: f"{row['protocol']}://{row['ip']}:{row['port']}", axis=1)

# Convert the series to a list
proxy_list = formatted_proxies.tolist()

# Example: Print the first 5 proxies
print(proxy_list[:5])


cl = Client()
login_timeout=0.1
# Function to perform login
def attempt_login(proxy):
    try:
        cl.set_proxy(proxy)
        print(f"Trying: {proxy}")
        login_response = cl.login("riley_simpson15", "Honk2020")
        if login_response:  # Assuming login_response is True for successful login
            print(f"Login successful with proxy {proxy}")
            return True
        else:
            print(f"Login failed with proxy {proxy}")
            return False
    except TimeoutError:
        #print(f"Connection timed out for proxy {proxy}, trying next one.")
        pass
    except Exception as e:
        #print(f"An error occurred for proxy {proxy}: {e}")
        return False

# Now, in your main loop:
for proxy in proxy_list:
    login_thread = threading.Thread(target=attempt_login, args=(proxy,))
    login_thread.start()
    login_thread.join(timeout=login_timeout)
    if login_thread.is_alive():
        #print(f"Login attempt timed out for proxy {proxy}, trying next one.")
        pass
    else:
        # Check the result of the login attempt
        if attempt_login(proxy):
            break  # Stop if login is successful
```
