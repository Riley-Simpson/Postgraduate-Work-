```python

import pandas as pd 
import sklearn 
import matplotlib.pyplot as plt
import numpy as np
```

```python

df = pd.read_csv("CS98XRegressionTrain.csv")
df.head()

converted_categorical = pd.get_dummies(df, columns=['top genre'], dtype=float).dropna()


# Define a function to categorize each year into its respective decade
def categorize_year_to_decade(year):
    return str(year)[:3] + "0's"


converted_categorical['decade']=converted_categorical['year'].apply(categorize_year_to_decade)
converted_categorical = converted_categorical.drop(columns=['year'])
converted_categorical = pd.get_dummies(converted_categorical, columns=['decade'], dtype=float)

def char_length(string):
    return len(string)

def word_count(word):
    return word.count(' ')+1

converted_categorical['title_length'] = converted_categorical['title'].apply(char_length)
converted_categorical['title_word_count'] = converted_categorical['title'].apply(word_count)
converted_categorical=converted_categorical.drop(columns=['title'])

converted_categorical['artist_length'] = converted_categorical['artist'].apply(char_length)
converted_categorical['artist_word_count'] = converted_categorical['artist'].apply(word_count)
converted_categorical=converted_categorical.drop(columns=['artist'])


converted_categorical.to_csv("Training_Cleaned.csv")
```
