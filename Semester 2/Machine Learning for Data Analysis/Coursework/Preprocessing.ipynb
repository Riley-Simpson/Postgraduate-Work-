```python

import pandas as pd 
import sklearn 
import matplotlib.pyplot as plt
import numpy as np
```

```python

df = pd.read_csv("CS98XRegressionTest.csv")
df.head()

converted_categorical = pd.get_dummies(df, columns=['top genre'], dtype=float).dropna()


# Define a function to categorize each year into its respective decade
def categorize_year_to_decade(year):
    return str(year)[:3] + "0's"


converted_categorical['decade']=converted_categorical['year'].apply(categorize_year_to_decade)
converted_categorical = converted_categorical.drop(columns=['year'])
converted_categorical = pd.get_dummies(converted_categorical, columns=['decade'], dtype=float)

def char_length(string):
    return len(string)

def word_count(word):
    return word.count(' ')+1

converted_categorical['title_length'] = converted_categorical['title'].apply(char_length)
converted_categorical['title_word_count'] = converted_categorical['title'].apply(word_count)
converted_categorical=converted_categorical.drop(columns=['title'])

converted_categorical['artist_length'] = converted_categorical['artist'].apply(char_length)
converted_categorical['artist_word_count'] = converted_categorical['artist'].apply(word_count)
converted_categorical=converted_categorical.drop(columns=['artist'])

converted_categorical=converted_categorical.set_index('Id')
converted_categorical.to_csv("Testing_Cleaned.csv", index=False)
converted_categorical.head()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the data
df = pd.read_csv("CS98XRegressionTest.csv")

# Initialize OneHotEncoder
encoder = OneHotEncoder(dtype=float)

# Fit and transform the 'top genre' column
top_genre_encoded = encoder.fit_transform(df[['top genre']]).toarray()

# Create a DataFrame from the encoded columns
top_genre_encoded_df = pd.DataFrame(top_genre_encoded, columns=encoder.get_feature_names_out(['top genre']))

# Concatenate the new DataFrame with the original DataFrame, dropping the original 'top genre' column
df = pd.concat([df.drop('top genre', axis=1), top_genre_encoded_df], axis=1).dropna()

# Define a function to categorize each year into its respective decade
def categorize_year_to_decade(year):
    return str(year)[:3] + "0's"

df['decade'] = df['year'].apply(categorize_year_to_decade)
df = df.drop(columns=['year'])

# Apply OneHotEncoding to the 'decade' column
decade_encoded = encoder.fit_transform(df[['decade']]).toarray()
decade_encoded_df = pd.DataFrame(decade_encoded, columns=encoder.get_feature_names_out(['decade']))
df = pd.concat([df.drop('decade', axis=1), decade_encoded_df], axis=1)

# Remaining transformations
def char_length(string):
    return len(string)

def word_count(word):
    return word.count(' ') + 1

df['title_length'] = df['title'].apply(char_length)
df['title_word_count'] = df['title'].apply(word_count)
df = df.drop(columns=['title'])

df['artist_length'] = df['artist'].apply(char_length)
df['artist_word_count'] = df['artist'].apply(word_count)
df = df.drop(columns=['artist'])

df = df.set_index('Id')
df.to_csv("Testing_Cleaned.csv", index=False)
df.info()
```
